```python
import thinkbayes
import thinkplot
```

```python
class MeasureHits(thinkbayes.Suite):
    
    def __init__(self, name=''):
        pmf = thinkbayes.MakeUniformPmf(0,100,101)
        thinkbayes.Suite.__init__(self, pmf, name=name)
        
    def Likelihood(self, data, hypothesis):
        """Computes the likelihood of the data under the hypothesis.
        
        Evaluates the Poisson PMF for lambda and k.
        
        hypothesis: rate of measure hits per hour
        data: measure hits in one hour
        """
        lam = hypothesis
        k = data
        like = thinkbayes.EvalPoissonPmf(k, lam)
        return like
```

```python
s = MeasureHits('firstspike')
thinkplot.Pmf(s)
```

```python
s.Update(10)
thinkplot.Pmf(s)
```
